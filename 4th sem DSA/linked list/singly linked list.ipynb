```python
class Node:
    def __init__(self,data):
        self.data = data
        self.ref = None
```

```python
class LinkedList:
    def __init__(self):
        self.head = None
    def traverse(self):
        n = self.head
        while n != None:
            print(n.data)
            n = n.ref
    def add_begin(self,data):
        new_node = Node(data)
        new_node.ref = self.head
        self.head = new_node
    
    def add_end(self,data):
        new_node = Node(data)
        if self.head != None:
            n = self.head
            while n.ref != None:
                n = n.ref
            n.ref = new_node
        else:
            self.head = new_node
    def add_middle(self,data,pos):
        new_node = Node(data)
        if pos == 0:
            new_node.ref = self.head
            self.head = new_node    
        else:
            n = self.head
            i = 1
            while i != pos-1 and n.ref != None:
                n = n.ref
                i += 1
            if i == pos-1:    
                temp1 = n.ref
                new_node.ref = temp1
                n.ref = new_node
    def delete_begin(self):
        if self.head != None:
            self.head = self.head.ref
        else:
            print('can"t delete from empty linked list')
    def delete_middle(self,pos):
        if self.head != None:
            if pos == 1:
                self.delete_begin()
                return
            i = 1
            n = self.head
            while i != pos-1 and n.ref != None:
                n = n.ref
                i += 1
            if i == pos-1:
                k = n.ref.ref
                n.ref = k
    def delete_end(self):
        if self.head != None:
            n = self.head
            while n.ref != None:
                n = n.ref
            n.ref = None
        else:
            print('can"t delete from empty linked list')
```

```python
ll = LinkedList()
l = int(input('enter the number of nodes:'))
print('1.add_begin\n2.add_end\n3.add_middle\n4.delete_begin\n5.delete_end\n6.delete_middle\n7.traverse')
i = 0
while i <= l:
    choice = int(input('enter choice'))
    if choice == 1:
        data = int(input('enter data:'))
        ll.add_begin(data)
        i += 1
    elif choice == 2:
        data = int(input('enter data:'))
        ll.add_end(data)
        i += 1
    elif choice == 3:
        data = int(input('enter data:'))
        pos = int(input('enter position:'))
        ll.add_middle(data,pos)
        i += 1
    elif choice == 4:
        ll.delete_begin()
        i -= 1
    elif choice == 5:
        ll.delete_end()
        i -= 1
    elif choice == 6:
        pos = int(input('enter position:'))
        ll.delete_middle(pos)
        i -= 1
    elif choice == 7:
        ll.traverse()
    else:
        print('enter proper choice!!!')
```
